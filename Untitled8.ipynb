```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext.getOrCreate();
spark = SparkSession(sc)
```

```python
pydf_1="/home/aa/Downloads/finaldata.csv"
```

```python
from pyspark.sql.types import*
```

```python
schema = StructType() \
            .add("sr.no",IntegerType(),True) \
            .add("track_id ",StringType(),True) \
            .add('track_name', StringType(), True) \
            .add("track_popularity",IntegerType(),True) \
            .add("duration_ms",IntegerType(),True) \
            .add("explicit",IntegerType(),True) \
            .add("release_date",IntegerType(),True) \
            .add("danceability",FloatType(),True) \
            .add("energy",FloatType(),True) \
            .add("key",IntegerType(),True) \
            .add("loudness",FloatType(),True) \
            .add("mode",IntegerType(),True) \
            .add("speechiness",FloatType(),True) \
            .add("acousticness",FloatType(),True) \
            .add("instrumentalness",FloatType(),True) \
            .add("liveness",FloatType(),True) \
            .add("valence",FloatType(),True) \
            .add("tempo",FloatType(),True) \
            .add("time_signature",IntegerType(),True) \
            .add("artists_name",StringType(),True) \
            .add("id_artists",StringType(),True) \
            .add("followers",IntegerType(),True) \
            .add("popularity",IntegerType(),True) \
            .add("genres",StringType(),True) 
```

```python
pydf_1=spark.read.csv(pydf_1, header=True, schema=schema)
```

```python
pydf_1.createOrReplaceTempView('tbl1')
```

```python
spark.sql("SELECT *FROM tbl1").show()
```

```python
#YEAR WISE POPULARITY OF THE SONG(FEATURE)

spark.sql("SELECT release_date,sum(popularity) as popularity,track_name FROM tbl1 GROUP BY release_date,track_name ORDER BY popularity DESC").show(100)
```

```python
#WHICH GENERES IS MOST POPULAR

spark.sql("SELECT genres,track_name,count(popularity) AS Total_Popularity FROM tbl1 GROUP BY genres,track_name ORDER BY Total_Popularity DESC").show(10)
```

```python
#WHICH GENRE HAS MOST FAMOUS SINGER BASED ON POPULARITY

spark.sql("SELECT genres,artists_name,sum(popularity) AS Total_Popularity FROM tbl1 GROUP BY genres,artists_name ORDER BY Total_Popularity DESC").show(1000)
```

```python
#WITH YEAR IS THEIR POPULARITY IN INSTRUMENTAL SONFGS

spark.sql("SELECT release_date,track_name ,instrumentalness,sum(popularity) AS Total_Popularity FROM tbl1 GROUP BY release_date,track_name,instrumentalness ORDER BY Total_Popularity DESC").show(10)
```

```python
#WHICH ARTIST HAVING MORE FOLLOWER

spark.sql("SELECT artists_name, followers FROM tbl1 ").show(1000)
```

```python
#AVERGAE INSTRUMENTALS IN THE TRACKS

spark.sql("SELECT AVG(instrumentalness) FROM tbl1").show()
```

```python
#WHICH YEAR WHICH GENRES IS MOST POPULAR

spark.sql("SELECT genres,sum(popularity) AS popu_artist,release_date from tbl1 GROUP BY genres,release_date ORDER BY popu_artist DESC").show(10)
```

```python
#Popular speechiness Track

spark.sql("SELECT track_name,speechiness,max(track_popularity) AS track_popularity FROM tbl1 GROUP BY track_name,speechiness ORDER BY track_popularity DESC").show(1000)
```

```python
#Track mode( major (1) or minor (0) )

spark.sql("""SELECT track_name,
                    mode,
                    CASE WHEN mode = 0 THEN 'MAJOR'
                        WHEN mode = 1 THEN 'MINOR'
                         ELSE 'N'END AS TRACK_MODE
              FROM tbl1""").show(20)
```

```python
#Popular loudness track

spark.sql("SELECT max(track_popularity) AS track_popularity,loudness,track_name FROM tbl1 GROUP BY loudness,track_name ORDER BY track_popularity  DESC ").show(10)
```

```python
#Popular loudness track

spark.sql("SELECT max(loudness) AS loudness,track_name FROM tbl1 GROUP BY track_name ORDER BY loudness  DESC ").show(10)
```

```python
#Popular energetic track

spark.sql("SELECT max(energy) AS energy,track_name FROM tbl1 GROUP BY track_name ORDER BY energy  DESC ").show(10)
```

```python
#Year wise time duration of track

spark.sql("SELECT track_name,max(duration_ms) AS duration_ms,artists_name,release_date FROM tbl1 GROUP BY track_name,release_date,artists_name ORDER BY duration_ms DESC").show(10)
```

```python
#Year wise time duration of track(correct)

spark.sql("SELECT track_name,duration_ms, release_date FROM tbl1 DESC").show(1000)
```

```python
#Year wise time duration of track

spark.sql("SELECT track_name,duration_ms, max(release_date) AS release_date FROM tbl1 GROUP BY track_name,duration_ms ORDER BY release_date DESC").show(1000)
```

```python
#Artist & it's time duration of track

spark.sql("SELECT track_name,sum(duration_ms) AS duration_ms,artists_name FROM tbl1 GROUP BY track_name,artists_name ORDER BY duration_ms DESC").show(10)
```
